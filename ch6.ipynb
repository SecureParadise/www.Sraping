```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import random
import datetime
```

```python
seed_value = int(datetime.datetime.now().timestamp())
random.seed(seed_value)
```

```python

def getLinks(articleUrl):
    html = urlopen('http://en.wikipedia.org{}'.format(articleUrl))
    bs = BeautifulSoup(html,'html.parser')
    return bs.find('div',{'id':'bodyContent'}).find_all('a',href = re.compile('^(/wiki/)(?!:).*$'))
```

```python
links = getLinks('/wiki/Kevin_Bacon')
```

```python
links
```

```python
len(links)
```

```python
while len(links) > 0:
    newArticle = links[random.randint(0,len(links)-1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
```

```python
def getExternalLinks(bs,url):
    internal_netloc =urlparse(url).netloc
    ExternalLinks = set()
    for link in bs.find_all('a'):
        if not link.attrs.get('href'):
            continue
        parsed = urlparse(link.attrs['href'])
        if parsed.netlock != '' and parsed.netlock != internal_netloc:
            ExternalLinks.add(link.attrs['href'])
    return list(ExternalLinks)

```
